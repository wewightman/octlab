```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bg = plt.imread('./data/tape_BG-RAW-20190516-172201.tif')
bg = np.mean(bg, axis = 0)
```

```python
raw = plt.imread('./data/tape_RAW-RAW-20190516-172228.tif')


plt.figure()
plt.hist(raw.flatten(), bins=200)
plt.show()

plt.figure()
plt.plot(raw[1,:])
plt.plot(bg)
plt.show()

plt.figure()
plt.plot((np.sin(np.pi*np.arange(len(bg))/len(bg)))* (raw[1,:]-bg))
plt.show()

plt.figure()
freq = np.abs(np.fft.fft((np.sin(np.pi*np.arange(len(bg))/len(bg)))* (raw[1,:]-bg)))
plt.plot(freq)
plt.show()
```
